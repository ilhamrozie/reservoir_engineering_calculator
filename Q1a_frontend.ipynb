```python
from tkinter import *
import Q1a_backend 


window = Tk()
window.title("Reservoir Engineering Calculator")

#create function for calc command that calls function from backend. this function will use user's input to output q(flowrate) 
def calc_flowrate_comm():
    k = float(e1.get())
    u = float(e2.get())
    A = float(e3.get())
    L = float(e31.get())
    deltap = float(e4.get())
    q = Q1a_backend.calc_flowrate(k, A, u, L, deltap)
    e5.delete(0, END)
    e5.insert(END, "{:.2f}".format(q))    

    
def calc_sw_comm():
    a = float(e7.get())
    por = float(e8.get())
    Rw = float(e9.get())
    Rt = float(e10.get())
    m = float(e11.get())
    n = float(e12.get())
    sw = ((a*Rw)/(Rt*(por**m)))**(1/n)
    e13.delete(0, END)
    e13.insert(END, "{:.2f}".format(sw))
    
    
def calc_m_comm():
    deltatfl = float(e15.get())
    deltat = float(e16.get())
    rhob = float(e17.get())
    rhofl = float(e18.get())
    m = ((deltatfl-deltat)/(rhob-rhofl))*0.01
    e19.delete(0, END)
    e19.insert(END, "{:.2f}".format(m))  
    
def calc_n_comm():
    poronfl = float(e21.get())
    poron = float(e22.get())
    rhob= float(e23.get())
    rhofl= float(e24.get())
    n = ((poronfl-poron)/(rhob-rhofl))
    e25.delete(0, END)
    e25.insert(END, "{:.2f}".format(n))  
    
    
    



#output a text
#darcy's law
l0 = Label(window, text="Flowrate Calculator (Linear Darcy's Law) ")
l0.grid(row=0, column=0)

l1 = Label(window, text='Permeability, k (md)')
l1.grid(row=1, column=0)

l2 = Label(window, text='Viscosity, \u03BC (cp) ')
l2.grid(row=1, column=2)

l3 = Label(window, text='Area open to flow, AOF (ft^2)')
l3.grid(row=2, column=0)

l4 = Label(window, text='Length open to flow, LOTF (ft)')
l4.grid(row=2, column=2)

l4 = Label(window, text='Differential pressure, \u0394p (psia)')
l4.grid(row=3, column=0)

l5 = Label(window, text='Flowrate, q (bpd)')
l5.grid(row=3, column=2)





#output a textbox for user input
perm_text = StringVar()
e1 = Entry(window, textvariable = perm_text)
e1.grid(row=1, column=1)

visc_text = StringVar()
e2 = Entry(window, textvariable =visc_text)
e2.grid(row=1, column=3)

AOF_text = StringVar()
e3 = Entry(window, textvariable = AOF_text)
e3.grid(row=2, column=1)

LOTF_text = StringVar()
e31 = Entry(window, textvariable = LOTF_text)
e31.grid(row=2, column=3)

deltap_text = StringVar()
e4 = Entry(window, textvariable = deltap_text)
e4.grid(row=3, column=1)

flowrate_text = StringVar()
e5 = Entry(window, textvariable = flowrate_text)
e5.grid(row=3, column=3)



#archie's equation
l6 = Label(window, text="Water Saturation Calculator (Archie's Equation)")
l6.grid(row=4, column=0)

l7 = Label(window, text="Tortuosity factor, a (-)")
l7.grid(row=5, column=0)

l8 = Label(window, text="Porosity, \u03C6 (-)")
l8.grid(row=6, column=0)

l9 = Label(window, text="Formation water resistivity, Rw (ohm-m)")
l9.grid(row=7, column=0)

l10 = Label(window, text="Total resistivity, Rt (ohm-m)")
l10.grid(row=8, column=0)

l11 = Label(window, text="Cementation exponent, m (-)")
l11.grid(row=5, column=2)

l12 = Label(window, text="Saturation exponent, n (-)")
l12.grid(row=6, column=2)

l13 = Label(window, text="SW Archie's, SwArch (-)")
l13.grid(row=7, column=2)



a_text = StringVar()
e7 = Entry(window, textvariable = a_text)
e7.grid(row=5, column=1)

poro_text = StringVar()
e8 = Entry(window, textvariable = poro_text)
e8.grid(row=6, column=1)

Rw_text = StringVar()
e9 = Entry(window, textvariable = Rw_text)
e9.grid(row=7, column=1)

Rt_text = StringVar()
e10 = Entry(window, textvariable = Rt_text)
e10.grid(row=8, column=1)

m_text = StringVar()
e11 = Entry(window, textvariable = m_text)
e11.grid(row=5, column=3)

n_text = StringVar()
e12 = Entry(window, textvariable = n_text)
e12.grid(row=6, column=3)

Swarch_text = StringVar()
e13 = Entry(window, textvariable = Swarch_text)
e13.grid(row=7, column=3)


#M parameter for lithology plot
l14 = Label(window, text="M Calculator for M-N Lithology Plot")
l14.grid(row=9, column=0)

l15 = Label(window, text="Fluid interval transit time, \u0394tfl (us/ft)")
l15.grid(row=10, column=0)

l16 = Label(window, text="Chosen interval transit time, \u0394t (us/ft)")
l16.grid(row=11, column=0)

l17 = Label(window, text="Bulk density, \u03A1B (g/cc)")
l17.grid(row=12, column=0)

l18 = Label(window, text="Fluid density, \u03A1l (g/cc)")
l18.grid(row=13, column=0)

l19 = Label(window, text="M (-)")
l19.grid(row=10, column=2)



tfl_text = StringVar()
e15 = Entry(window, textvariable = tfl_text)
e15.grid(row=10, column=1)

t_text = StringVar()
e16 = Entry(window, textvariable = t_text)
e16.grid(row=11, column=1)

rhob_text = StringVar()
e17 = Entry(window, textvariable = rhob_text)
e17.grid(row=12, column=1)

rhofl_text = StringVar()
e18 = Entry(window, textvariable = rhofl_text)
e18.grid(row=13, column=1)

m_text = StringVar()
e19 = Entry(window, textvariable = m_text)
e19.grid(row=10, column=3)


#N parameter for lithology plot
l20 = Label(window, text="N Calculator for M-N Lithology Plot")
l20.grid(row=14, column=0)

l21 = Label(window, text="Porosity of open space, \u03C6nfl (-)")
l21.grid(row=15, column=0)

l22 = Label(window, text="Neutron porosity, \u03C6n (-)")
l22.grid(row=16, column=0)

l23 = Label(window, text="Bulk density, \u03A1b (g/cc)")
l23.grid(row=17, column=0)

l24 = Label(window, text="Fluid density, \u03A1fl (g/cc)")
l24.grid(row=18, column=0)

l25 = Label(window, text="N (-)")
l25.grid(row=15, column=2)


phinfl_text = StringVar()
e21 = Entry(window, textvariable = phinfl_text)
e21.grid(row=15, column=1)

phin_text = StringVar()
e22 = Entry(window, textvariable = phin_text)
e22.grid(row=16, column=1)

rhob1_text = StringVar()
e23 = Entry(window, textvariable = rhob1_text)
e23.grid(row=17, column=1)

rhofl1_text = StringVar()
e24 = Entry(window, textvariable = rhofl1_text)
e24.grid(row=18, column=1)

n_text = StringVar()
e25 = Entry(window, textvariable = n_text)
e25.grid(row=15, column=3)




b1 = Button(window, text='Calculate flowrate', width=20, command=calc_flowrate_comm)
b1.grid(row=0, column=1)

b2 = Button(window, text='Calculate water saturation', width=20, command=calc_sw_comm)
b2.grid(row=4, column=1)

b3 = Button(window, text='Calculate M', width=20, command=calc_m_comm)
b3.grid(row=9, column=1)

b4 = Button(window, text='Calculate N', width=20, command=calc_n_comm)
b4.grid(row=14, column=1)



window.mainloop()
```
